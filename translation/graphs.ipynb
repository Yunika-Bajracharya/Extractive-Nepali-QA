```python
import pandas as pd
import pickle

with open('data.pickle', 'rb') as file:
    data = pickle.load(file)
```

```python
found_lens = list(map(lambda x: len(x), data['found']))
not_found_lens = list(map(lambda x: len(x), data['not_found']))
```

```python
import numpy as np
median_value = np.median(found_lens)

# Print the result
print("Median:", median_value)
median_value = np.median(not_found_lens)

# Print the result
print("Median:", median_value)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
found_lengths = list(map(lambda x: len(x), data['found']))
not_found_lengths = list(map(lambda x: len(x), data['not_found']))
found_lengths = list(filter(lambda x: x < 120, found_lengths))
not_found_lengths = list(filter(lambda x: x < 120, not_found_lengths))

fig, ax = plt.subplots(figsize = (13,5))
sns.set(rc={'axes.labelsize': 16, 'ytick.labelsize': 14, 'xtick.labelsize': 14})
sns.kdeplot(found_lengths, fill=True, label='Accepted')
sns.kdeplot(not_found_lengths, fill=True, label='Discarded')
ax.set_title("Densities of answer lengths using Naive approach")
ax.set_xlabel("Length of answer")
ax.set_ylabel("Density")
ax.legend()
plt.savefig('histogram_plot.png')
plt.show()
```

```python
import pandas as pd

df1 = pd.read_csv('translated2/combined_excel_1-translated.csv')
df2 = pd.read_csv('translated2/combined_excel_2-translated.csv')
df3 = pd.read_csv('translated2/combined_excel_3-translated.csv')
df4 = pd.read_csv('translated2/combined_excel_4-translated.csv')
```

```python
df = pd.concat([df1, df2, df3, df4])
new_method_lens = list(map(lambda x: len(x), df.iloc[:, 2]))
```

```python
df.head(20)
```

```python
# loop through df row by row
doublicate_count = 0
no_answer_count = 0
for index, row in df.iterrows():
    # row.iloc[0] = row.iloc[0].replace('*', '')
    # row.iloc[0] = row.iloc[0].strip()

    context = row.iloc[0]
    question = row.iloc[1]
    answer = row.iloc[2]

    # python find number of occurences of a substring in a string
    answer_count = context.count(answer)
    if answer_count > 1:
        doublicate_count += 1
    if answer_count == 0:
        no_answer_count += 1
    
print("doublicate_count:", doublicate_count)
print("no_answer_count:", no_answer_count)
```

```python
print("Mean: ", np.mean(new_method_lens))
print("Median: ", np.median(new_method_lens))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize = (13,5))

sns.set(rc={'axes.labelsize': 16, 'ytick.labelsize': 14, 'xtick.labelsize': 14})
filtered_new_method_lens = list(filter(lambda x: x < 120, new_method_lens))
sns.kdeplot(found_lengths, fill=False, label='Accepted')
sns.kdeplot(not_found_lengths, fill=False, label='Discarded')
sns.kdeplot(filtered_new_method_lens, fill=True, label='New method', color='green')


ax.set_title("By placing special tokens(**) around answers")
ax.set_xlabel("Length of answer")
ax.set_ylabel("Density")
ax.legend()
plt.savefig('histogram_plot.png')
plt.show()
```
